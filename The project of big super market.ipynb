# Visual for each componant of the Features that effects the profit 

```python
# The feature of the city 
print(df['City'].value_counts())
sns.barplot(x='City',y='gross income',data=df)
'''from the below results we get we need to make more marketing in Yangon yo get more income'''
```

### Which Branch is the most active and get more audiance and how is the least

```python
#the feature of Branch
sns.barplot(x='Branch',y='gross income',data=df)
print(df['Branch'].value_counts())
'''from the below results we get we need to make more marketing in Branch A yo get more income'''
```

### which Gender is more paid ?

```python
#the feature of Gender
sns.barplot(x='Gender',y='gross income',data=df)
print(df['Gender'].value_counts())
'''from the below results we get we need to make more marketing to male get more income and make more offers to the female'''
```

```python
sns.boxplot(x='Gender',y='Rating',data=df)
```

### Which way of Payment is more comman ?

```python
#the feature of Payment
sns.barplot(x='Payment',y='gross income',data=df)
print(df['Payment'].value_counts())
'''from the below results we get we need to make more offers on Credit card and Ewallet get more income and anther note that Ewallet is more in time of usage but the value of profit is lower than what should be'''
```

### Which Product line make the max of Gross income

```python
#the feature of Product line
sns.barplot(x='Product line',y='gross income',data=df)
plt.xticks(rotation=90)
print(df['Product line'].value_counts())
'''from the below results we get we need to make more marketing in Fashion accessories to get more income and make sure about the value of this feature although it is the biggest type of sales but it is the lowest value with respect to other and (Home and lifestyle ) is the heighst profit income of the total buying power '''
```

```python
sns.boxplot(x='Product line',y='Rating',data=df)
plt.xticks(rotation=90)
```

```python
sns.boxplot(y='Tax 5%',x='Product line',data=df)
plt.xticks(rotation=90)
```
